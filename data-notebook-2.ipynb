```python
import numpy as np
import pandas as pd
```

```python
X = np.load('/kaggle/input/data-notebook/test_imgs.npy')
y = np.load('/kaggle/input/data-notebook/test_msks.npy')
```

```python
def shuffle_images_and_masks(images, masks):
    
    # Generate a random permutation of indices
    permutation = np.random.permutation(len(images))
    
    # Use advanced indexing to shuffle both arrays
    shuffled_images = images[permutation]
    shuffled_masks = masks[permutation]
    
    return shuffled_images, shuffled_masks
```

```python
tempX, tempy = shuffle_images_and_masks(X, y)
```

```python
newx = tempX[:10000]
newy = tempy[:10000]
```

```python
np.save('/kaggle/working/Xtrain.npy', newx)
np.save('/kaggle/working/ytrain.npy', newy)
```
