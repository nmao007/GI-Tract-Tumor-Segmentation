```python
import numpy as np
import pandas as pd
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
X = np.load("/kaggle/input/datanotebook-2/Xtrain.npy")
y = np.load("/kaggle/input/datanotebook-2/ytrain.npy")
```

```python
def is_mask_blank(mask):
    # Check if all pixels are zero (or below a small threshold)
    threshold = 1e-7  # Use a small threshold to account for potential floating-point imprecision
    return np.all(mask < threshold)
```

```python
present_imgs = []
present_msks = []
blank_imgs = []
blank_msks = []

for i in range(len(y)):
    if (is_mask_blank(y[i]) == False):
        present_imgs.append(X[i])
        present_msks.append(y[i])
    else:
        blank_imgs.append(X[i])
        blank_msks.append(y[i]) 
```

```python
len(present_msks)
```

```python
present_imgs = present_imgs[:3200]
present_msks = present_msks[:3200]
```

```python
blank_imgs = blank_imgs[:800]
blank_msks = blank_msks[:800]
```

```python
X_combined = np.concatenate((present_imgs, blank_imgs), axis=0)
y_combined = np.concatenate((present_msks, blank_msks), axis=0)
```

```python
print(len(X_combined))
len(y_combined)
```

```python
indices = np.arange(len(X_combined))

# Shuffle the indices
np.random.shuffle(indices)

# Use these indices to shuffle both X and y consistently
X_shuffled = X_combined[indices]
y_shuffled = y_combined[indices]
```

```python
X = np.array(X_shuffled)
y = np.array(y_shuffled)
```

```python
np.save('X.npy', X)
np.save('y.npy', y)
```
